```python
import os
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import numpy as np 
```

```python
df = pd.read_csv("clothes_price_data.csv")
```

```python
# Line chart
plt.figure(figsize=(5, 3))
try:
    sns.lineplot(data=df, x='ItemName', y='Price')
    plt.title("Item Name Over Price")
    plt.xlabel("Item Name")
    plt.ylabel("Price")
    plt.show()
except KeyError:
    print("Error: Ensure 'Item Name' and 'Price' are present in the dataset.")
```

```python
# Bar chart
plt.figure(figsize=(5, 3))
try:
    sns.barplot(x='Category', y='Price', data=df, ci=None)
    plt.title("Price Across Categories")
    plt.xlabel("Category")
    plt.ylabel("Price")
    plt.show()
except KeyError:
    print("Error: Ensure 'Category' and 'Price' are present in the dataset.")
```

```python
# Histogram
plt.figure(figsize=(5, 3))
try:
    sns.histplot(df['Price'], kde=True, bins=20)
    plt.title("Distribution of Price")
    plt.xlabel("Price")
    plt.ylabel("Frequency")
    plt.show()
except KeyError:
    print("Error: Ensure 'Price' is present in the dataset.")
```
